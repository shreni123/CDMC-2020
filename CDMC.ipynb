```python
#Accuracy
def acc(arg):
    sum=0
    for i in range(0,12):
        if arr[i]>arg[i]:
            sum=sum+(arg[i]/arr[i])
        else:
            sum=sum+(arr[i]/arg[i])
    return (sum*100/12)
        
```
